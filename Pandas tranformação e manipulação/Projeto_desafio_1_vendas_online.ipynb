```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Leitura e visualização dos dados
dados = pd.read_json('/content/dados_vendas_clientes.json')
dados.head()
```

```python
# Aplicando o metodo normalize na coluna 'dados_vendas'
dados = pd.json_normalize(dados['dados_vendas'])
dados
```

```python
# Armazenar as colunas em uma lista, para utiliar o metodo explode
colunas = list(dados.columns)
```

```python
# Utilizar o metodo explode nas colunas desejadas
dados = dados.explode(colunas[1:])
```

```python
# Visualizar o df para verificar as alterações feitas
dados
```

```python
# Resetar os indices do df apos o explode
dados.reset_index(inplace=True, drop=True)
dados
```

```python
# Verificar os tipos de dados do df
dados.info()
```

```python
# Remover os caracteres e virgulas para poder fazer a conversão dos dados
dados['Valor da compra'] = dados['Valor da compra'].apply(lambda x : x.replace('R$', '').replace(',', '').strip())
```

```python
# Importar a biblioteca numpy, para a conversão dos dados
import numpy as np
```

```python
# Converter os valores da coluna para float64
dados['Valor da compra'] = dados['Valor da compra'].astype(np.float64)
```

```python
# Verificar a conversão do tipo de dado
dados.info()
```

```python
dados['Cliente']
```

```python
dados['Cliente'] = dados['Cliente'].str.lower()
dados['Cliente']
```

```python
# Regex não seleciona as letras minusculas e espaços em branco
# Todo o restante é apagado
dados['Cliente'] = dados['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()
dados.head()
```

```python
# Converter os dados para o tipo datetime e verificar a conversão
dados['Data de venda'] = pd.to_datetime(dados['Data de venda'], format='%d/%m/%Y')
dados['Data de venda']
```

```python
# Agrupar os clientes e somar os valores de compras totais de suas compras e classificar do maior para o menor
total_compras = dados.groupby(['Cliente'])['Valor da compra'].sum().sort_values(ascending=False)
total_compras
```

Diante das análises feitas com a base de dados, podemos perceber que durante o período de duração do evento o(a) cliente que realizou o maior valor de compras foi **Isabely Joanes**
