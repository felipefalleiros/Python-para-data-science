```python
# Importar a biblioteca pandas
import pandas as pd
```

```python
# Ler e visualiar os dados
dados = pd.read_json('/content/dados_locacao_imoveis.json')
dados.head()
```

```python
# Aplicar o metodo normalie na coluna 'dados_locacao' e visualizar os dados 'normalizados'
dados = pd.json_normalize(dados['dados_locacao'])
dados
```

```python
# Guardar as colunas em uma lista
colunas = list(dados.columns)
colunas
```

```python
# Realizar o método explode nas colunas desejadas
dados = dados.explode(colunas[1:])
dados
```

```python
# Resetar os índices do df após o explode
dados.reset_index(inplace=True, drop=True)
dados
```

```python
# Verificar os tipos de dados
dados.info()
```

```python
# Tratar valores para fazer a alteração para o tipo númerico
dados['valor_aluguel'] = dados['valor_aluguel'].apply(lambda x : x.replace('$', '').replace(',', '').replace('reais', '').strip())
```

```python
# Verificar alterações
dados
```

```python
import numpy as np
```

```python
# Converter os dados para o tipo númerico
dados['valor_aluguel'] = dados['valor_aluguel'].astype(np.float64)
```

```python
dados
```

```python
# Verificar a conversão de tipo
dados.info()
```

```python
# Substituir a expressão (blocoAP), fazendo uso da \ para considerar os parênteses
dados['apartamento'] = dados['apartamento'].str.replace(' \(blocoAP\)', '', regex=True)
dados
```

```python
# Converter para datatime as colunas que apresentam datas
dados['datas_combinadas_pagamento'] = pd.to_datetime(dados['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados['datas_de_pagamento'] = pd.to_datetime(dados['datas_de_pagamento'], format='%d/%m/%Y')

dados.info()
```

```python
# Adicionar uma nova coluna contendo os dias de atraso entre as datas de pagamentos e as datas de pagamento combinadas
dados['dias_atraso'] = (dados['datas_de_pagamento'] - dados['datas_combinadas_pagamento']).dt.days
dados
```

```python
# Tempo medio de atraso de pagamento de aluguel por apartamento
media_atraso = dados.groupby(['apartamento'])['dias_atraso'].mean().sort_values(ascending=False)
media_atraso
```

Através da análise da base de dados tivemos como resultado o tempo médio do atraso de pagamento de aluguel de cada apartamento presente no condomínio. Podendo através disso propor uma atualização das datas de pagamento combinadas, para assim evitar atrasos no futuro.
